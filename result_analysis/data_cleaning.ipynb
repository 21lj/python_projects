```python
import pandas as pd
import csv
import re
import os
from google.colab import files
```

```python
print("üìÇ Please upload your Excel (.xls or .xlsx) file:")
uploaded = files.upload()

for filename in uploaded.keys():
    INPUT_FILE = filename
```

```python
temp_csv = "temp_converted.csv"
try:
    df = pd.read_excel(INPUT_FILE)
    df.to_csv(temp_csv, index=False)
    print(f"‚úÖ Converted '{INPUT_FILE}' ‚Üí '{temp_csv}'")
except Exception as e:
    print(f"‚ùå Error reading Excel file: {e}")
    raise
```

```python
OUTPUT_FILE = "BCA_S1_results_cleaned.csv"

CLEAN_HEADERS = [
    'Slno',
    'Register_Number',
    'Name',
    'Digital_Marks',
    'Digital_Grade',
    'Maths_Marks',
    'Maths_Grade',
    'C_Marks',
    'C_Grade',
    'Lab_Marks',
    'Lab_Grade',
    'Cyber_Laws_Marks',
    'Cyber_Laws_Grade',
    'English_Marks',
    'English_Grade',
    'Language_Marks',
    'Language_Grade',
    'Failed_Subjects',
    'Total',
    'Percentage',
    'Overall_Grade'
]
```

```python
def clean_csv(input_csv, output_csv):
    print(f"\nüßπ Starting cleanup of '{input_csv}'...")
    cleaned_data = []

    try:
        with open(input_csv, mode='r', encoding='utf-8') as infile:
            reader = csv.reader(infile)

            # Skip the first 3 header rows
            for _ in range(3):
                next(reader, None)

            # Process each student row
            for row in reader:
                # Skip empty or malformed rows
                if not row or len(row) < 22:
                    continue

                # Skip discontinued students
                name_field = row[3]
                if 'DISCONTINUED' in name_field:
                    continue

                # Clean name field (remove DOB text)
                cleaned_name = re.sub(r'\s*\(.*DOB.*\)', '', name_field).strip()

                try:
                    # Slice columns 1‚Äì21 (21 fields expected)
                    new_row = row[1:22]

                    if len(new_row) != 21:
                        print(f"‚ö†Ô∏è Skipping row with {len(new_row)} fields instead of 21: {row}")
                        continue

                    # Replace Name field (index 2)
                    new_row[2] = cleaned_name

                    cleaned_data.append(new_row)

                except IndexError:
                    print(f"‚ö†Ô∏è Skipping malformed row: {row}")
                    continue

        # Write cleaned data
        with open(output_csv, mode='w', encoding='utf-8', newline='') as outfile:
            writer = csv.writer(outfile)
            writer.writerow(CLEAN_HEADERS)
            writer.writerows(cleaned_data)

        print(f"‚úÖ Successfully cleaned data and saved to '{output_csv}'")
        print(f"üë©‚Äçüéì Total students processed: {len(cleaned_data)}")

    except Exception as e:
        print(f"‚ùå Error while cleaning: {e}")
```

```python
clean_csv(temp_csv, OUTPUT_FILE)

files.download(OUTPUT_FILE)
print("\nüì• Cleaned file ready for download!")
```
